**Creating meaningful measures**

This company has daily laboratory tasks, from where a few samples can be collected. So I decided to create a measure named 'Sampeling' which states the actual numbers of samples collected from the number of total tasks. 

To further see the quality of this collected samples, I created the 'Defects%' and 'Fatal Error%' measures, which intuitively show the percentage of faulty, respectively fully damaged samples. 

Let me show you the three measures:

The actual formulas used to calculate these measures: (simple division from the total number of tasks, respectively samples)

Sampling% = DIVIDE(SUM('Quality Data'[Samples]), SUM('Quality Data'[Total Task]))

Defects% = DIVIDE(SUM('Quality Data'[Defects]),sum('Quality Data'[Samples]))

Fatal error% = DIVIDE(sum('Quality Data'[Fatel Errors]), sum('Quality Data'[Samples]))
