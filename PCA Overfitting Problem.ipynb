```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../DATA/matrice_PCs_ for RCT response.csv')
```

```python
df
```

```python
df['Target'] = df['Class'].map({'R' :1, 'NR':0})

#Aici pur si simplu am schimbat data type-ul target clasului in numere (litere nu se accepta in modele de ML)
```

```python
df
```

```python
df = df.drop('Class',axis =1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Target', axis = 1).values

y = df['Target'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

**In acest caz am considerat ca desi dataframe-ul nostru este extrem de mic, reviewer-ul probabila considerat faptul ca ati folosit cross-validation ca si overfitting. Motivul este ca metoda leave one out se foloseste de toti parametrii pentru antrenarea modelului, si dupa face preziceri pe ce s-a antrenat, si aici se pacaleste singur si face categoric overfitting. IN acest sens am folosit 20% din datele noastre alese in mod random pentru validare. Acest lucru va face ca training-ul sa fie extrem de zgomotos, dar validation sa existe intr-o masura ok.**   

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

#Aici am scalat datele, pentru ca modelele de deep learning/machine learning dau rezultate mai bune in acest mod 
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_test.shape
```

```python
#Acesta este un mdel extrem de simplu, secvential de deep learning. 
#Deep learning inseamna pur simplu o retea neuronala cu cel putin 2 layere inainte de layerul de output

model = Sequential()

model.add(Dense(10, activation = 'relu'))
model.add(Dropout(0.2))

model.add(Dense(4, activation = 'relu'))
model.add(Dropout(0.2))

#output neuron:

model.add(Dense(units = 1, activation = 'sigmoid'))

#Fiind o clasificare am folosit functia sigmoidala de activare

model.compile(optimizer = 'adam', loss = 'binary_crossentropy')
```

```python
model.fit(x = X_train, y = y_train, epochs = 500, 
         validation_data = (X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history).plot()
```

**In acest grafic se vede bine faptul ca atat loss (pierdera pe training) cat si val loss (cea pe test/validare) este in constanta coborare, deci nici o urma de overfitting.** 

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
predictions = (model.predict(X_test) > 0.45).astype("int32")
```

```python
print(classification_report(y_test, predictions))
```

Aici putem observa ca modelul nu functioneaza extrem de bine pentru R (1), dar din cate am inteles eu NR(0) reprezinta clasa de interes, iar pentru aceasta are un F1-score de 80% cu o precizie de 100%, deci modelul este unul reusit.

```python
confusion_matrix(y_test, predictions)
```

```python
import seaborn as sns
plt.figure(figsize=(12,4))
sns.heatmap(confusion_matrix(y_test, predictions), annot = True)
```

Acest lucru e observa si in cazul matricei de confuzie, unde la clasa NR nu a clasificat nici una gresit din cele 4 de test. In cazul R, a clasificat bine 2 din 3.
